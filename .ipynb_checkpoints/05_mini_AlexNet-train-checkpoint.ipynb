```python
train_loader, valid_loader = get_train_valid_loader(
    data_dir='./data/', batch_size=128, random_seed=1, 
    augment=False, download=True, subsample_frac=0.10
)
```

# Train - no pruning

```python
num_classes = 10 # output number of classes
num_epochs = 20  # number of total training loops?
# batch_size = 128  # how much data to train on at a time
learning_rate = 0.005 # SGD step size. Scheduler with 0.1 for 0-100, 0.01 for 100-200, and 0.001 for 200 to 250

model = MiniAlexNet(num_training_iter=num_epochs, num_classes=num_classes, verbose=False).to(device)
```

```python
train_losses_epoch, val_losses_epoch, test_df = full_train(
    model,
    train_loader, valid_loader, test_loader,
    learning_rate = 0.005,
    loss_fn = nn.CrossEntropyLoss(),
    model_update_params=False,
    plot=False,
    verbose=False,
)
```

```python
train_losses_df = pd.DataFrame.from_dict(train_losses_epoch)
val_losses_df = pd.DataFrame.from_dict(val_losses_epoch)
```

```python
training_losses_df = pd.DataFrame(train_losses_epoch).T
training_losses_df.columns.name = 'batch'
training_losses_df.index.name = 'epoch'

val_losses_df = pd.DataFrame(val_losses_epoch).T
val_losses_df.columns.name = 'batch'
val_losses_df.index.name = 'epoch'

stack_training_losses_df = training_losses_df.stack().reset_index().rename(columns={0: 'loss'})
stack_training_losses_df.index = stack_training_losses_df['epoch'].astype(str) + '-' + stack_training_losses_df['batch'].astype(str)
stack_training_losses_df.index.name = 'index'
stack_val_losses_df = val_losses_df.stack().reset_index().rename(columns={0: 'loss'})
stack_val_losses_df.index = stack_val_losses_df['epoch'].astype(str) + '-' + stack_val_losses_df['batch'].astype(str)
stack_val_losses_df.index.name = 'index'
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.scatterplot(stack_training_losses_df.reset_index(), x='index', y='loss', label='Training loss', alpha=0.5, c='gray', s=2)
sns.lineplot(stack_val_losses_df.reset_index(), x='index', y='loss',  label='Validation loss', c='red')
plt.xticks([])
plt.xlabel('Training and validation epoch/batch')
plt.show()
```

```python
sns.scatterplot(test_df.reset_index(), x='epoch', y='test_err', c='k')
plt.ylim(0, 1.0)
plt.ylabel("Test accuracy")
plt.show()
```
